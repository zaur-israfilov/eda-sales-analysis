```python
import pandas as pd

df = pd.read_csv("../data/sales_data.csv")

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df["total_bill"], bins=20)
plt.title("Distribution of Total Bill")
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.title("Tip vs Total Bill")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=df)
plt.title("Total Bill by Day")
plt.show()
```

```python
sns.boxplot(x="smoker", y="tip", data=df)
plt.title("Tip by Smoking Status")
plt.show()
```

```python
sns.boxplot(x="time", y="total_bill", data=df)
plt.title("Lunch vs Dinner Bills")
plt.show()
```

```python
df.groupby("day")["total_bill"].mean()
```

```python
df.groupby("time")["total_bill"].mean()
```

```python
df.groupby("smoker")["tip"].mean()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

# Exploratory Data Analysis – Restaurant Sales

This project explores a restaurant dataset to understand customer behavior, bill distribution, and tipping patterns.

## Data Loading
Load the dataset and inspect the first rows.

```python
import pandas as pd

df = pd.read_csv("../data/sales_data.csv")
df.head()
```

## Dataset Overview

We examine the dataset structure including rows, columns, and data types.

```python
df.shape
df.info()
```

## Statistical Summary

We analyze the main statistical properties of numeric variables.

```python
df.describe()
```

## Distribution of Total Bills

```python
sns.histplot(df["total_bill"], bins=20)
plt.title("Distribution of Total Bill")
plt.show()
```

## Tip vs Total Bill

```python
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.title("Tip vs Total Bill")
plt.show()
```

## Bills by Day

```python
sns.boxplot(x="day", y="total_bill", data=df)
plt.title("Total Bill by Day")
plt.show()
```

# Key Insights

• Weekend days generate higher restaurant revenue.

• Dinner transactions produce higher bills compared to lunch.

• Larger bills tend to result in higher tips.

• There is no significant tipping difference between smokers and non-smokers.
